K-Means Clustering

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
fd=norm_func(EW1.iloc[:,1:])
fd
```

```python
wcss=[]
for i in range(1,16):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(fd)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,16),wcss)
plt.title('Elbow Curve')
plt.xlabel('clusters')
plt.ylabel('wcss')
```

```python
#from Elbow Curve we have to decide 8 clusters
m1=KMeans(n_clusters=8)
m1.fit(fd)
```

```python
m1.labels_
```

```python
md=pd.Series(m1.labels_)
EW1['cluster']=md
EW1
```

```python
EW1.iloc[:,1:11].groupby(EW1.cluster).mean()
```

DBSCAN EastWest

```python
from sklearn.cluster import DBSCAN
```

```python
EW=pd.DataFrame()
xlfname='https://excelr.s3.nl-ams.scw.cloud/DataScience-Assignments/Assignments/clusters/EastWestAirlines.xlsx'
xl=pd.ExcelFile(xlfname)

for sheet in xl.sheet_names:
  df_tmp=xl.parse(sheet)
EW=EW.append(df_tmp,ignore_index=True,sort=False)
csvfile='sample.csv'
EW.to_csv(csvfile,index=False)
```

```python
EW.info()
```

```python
EW.drop(['ID#'],axis=1,inplace=True)
```

```python
array=EW.values
array
```

```python
from sklearn.preprocessing import StandardScaler
stscaler=StandardScaler().fit(array)
x=stscaler.transform(array)
```

```python
x
```

```python
dbs=DBSCAN(eps=2,min_samples=12)
dbs.fit(x)
```

```python
y=dbs.fit_predict(x)
```

```python
dbs.labels_
```

```python
cl=pd.DataFrame(y,columns=['cluster'])
cl
```

```python
pd.concat([EW,cl],axis=1)
```
